```python
#create a tree
import Queue

class Node:
    def __init__(self,value,parent,depth):
        self.value=value
        self.parent=parent                       #create a node
        self.depth=depth
parent=Node(2,None,0)
front=Queue.Queue()
front.put(parent)
while front:
    current_node=front.get()
    if current_node>3:
        break
    front.put(Node(current_node.value-1, current_node, current_node.depth+1))  #bfs
    front.put(Node(current_node.value+1, current_node, current_node.depth+1))
    print current_node.value
    
```

```python
import Queue
class Node:
    def __init__(self,value,parent,depth):
        self.value=value
        self.parent=parent                       #create a node
        self.depth=depth
parent=Node(10,None,0)
front=[]
front.append(parent)
while front:
    current_node=front.pop()
    child1=Node(current_node.value-1, current_node, current_node.depth+1)  #dfs
    child2=Node(current_node.value+1, current_node, current_node.depth+1)
    if child1.depth<=3:
        front.append(child2)
        front.append(child1)
    print current_node.value
```
